Scatter Plot Bananna :
-----------------------
Ab hum seekhenge scatter plot kaise banate hain.
Scatter plot ek graph hai jisme data points ko dots ke roop mein dikhaya
jata hai—yeh relationships ya patterns dikhane ke liye best hai.
..............


Kyun Zaroori Hai?
------------------

Jab tumhe do cheezon ke beech relation dekhna ho—like study hours aur marks ka
connection—tab scatter plot use hota hai. Machine learning mein features ke correlation check karne ke liye kaam aata hai.
.................


Kaise Karte Hain?
-------------------

plt.scatter() function use karte hain.
X-axis aur Y-axis pe values dete hain, aur har pair ek dot ban jata hai.

```python
# Data
hours = [2, 4, 6, 8, 10]
marks = [60, 75, 85, 90, 20]

# Scatter plot banayein
plt.scatter(hours, marks)

# Labels aur title add karo
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Marks vs Study Hours')

# Graph dikhayein
plt.show()
```

Scatter Plot Se Kya Extra Fayda Hai?
-----------------------------------------
Scatter plot ka bada fayda yeh hai ki yeh do cheezon ke beech ka sambandh (relationship) dikhata hai, jo line plot ya bar plot se directly nahi dikhta. Yeh batata hai ki ek cheez badalne se doosri pe kya asar hota hai. Chalo, isko samajhte hain:

----------------------------------------
Patterns Dikhata Hai:
------------------
Maan lo tumhare paas study hours aur marks hain. Scatter plot se tum ek nazar mein dekh sakte ho ki zyada hours padhne se marks badhte hain ya nahi.
Agar dots ek seedhi line jaisa pattern banate hain (upar ki taraf), toh matlab zyada study = zyada marks (positive relation).


Outliers Pata Chalte Hain:
---------------
Agar ek student ne 8 ghante padha par marks sirf 50 aaye, jabki baaki sabke 80+ hain, toh woh dot alag dikhega. Yeh outliers dikhane mein scatter plot best hai.


Correlation Check Karna:
-----------------
Machine learning mein tumhe pata karna hota hai ki koi feature (jaise hours) target (jaise marks) se kitna related hai. Scatter plot se yeh visually samajh aata hai—agar dots scattered (bikhre hue) hain, toh koi strong relation nahi; agar ek direction mein hain, toh relation hai.

```python

days = [1, 2, 3, 4, 5]
sales = [20, 25, 30, 35, 5]  # Last value outlier hai

plt.scatter(days, sales)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Daily Sales with Outlier')
plt.show()
```

Outliers Kya Hote Hain?
------
Outliers woh data points hote hain jo baaki data se kaafi alag hote hain—jaise bheed mein ek aadmi jo sabse lamba ya sabse chhota ho. Matlab, woh normal pattern ya trend ko follow nahi karte.

Simple Bhasha Mein:
--
Agar sab students 80-90 marks laa rahe hain, aur ek student 20 marks laaya, toh woh 20 wala "outlier" hai—kyunki woh baaki sab se bohot alag hai.


Scatter Plot Mein Outliers Kyun Dikhte Hain?
---
Scatter plot mein har data point ek dot hota hai. Jab baaki dots ek pattern banate hain (jaise ek line ya group), aur ek dot usse bohot door hota hai, toh woh outlier kehlata hai. Yeh baaki graphs (jaise bar ya line) se zyada clear scatter mein dikhta hai.

eg.
-
- 10 hours padhne ke baad 30 marks aana normal nahi lagta—shayad uss student ne padha nahi, ya test mein galti ki, ya data entry mein error hua.


Outliers Ka Fayda Ya Nuksaan:
--
Fayda:
-----------
Yeh dikhata hai ki data mein kuch unusual hai—machine learning mein yeh model ke liye problem ban sakta hai, toh pehle se pata chal jata hai.
- Real life mein: Ek shop mein sales har din 50-60 hai, par ek din 5 ho gayi—outlier se pata chalega ki kuch galat hua (jaise shop band tha).
  
Nuksaan:
-------

Agar outliers ko ignore na kiya jaye, toh model galat predict kar sakta hai—kyunki woh average ya trend ko disturb karte hain.

Colors Aur Size Customize Karna
---
Ab hum seekhenge ki scatter plot (ya kisi bhi plot) mein colors aur sizes kaise change karte hain. Yeh graph ko zyada attractive aur meaningful banata hai.

Kyun Zaroori Hai?
--

- Colors se alag-alag categories ya importance highlight kar sakte ho—like pass/fail students ko alag rang dena.
- Size se data ki intensity ya value dikhayi ja sakti hai—like zyada sales wale points bade dots se.
- Machine learning mein, yeh features ke differences ko visually samajhne mein madad karta hai.

Kaise Karte Hain?
--

- color: Plot ka rang badalne ke liye (jaise 'red', 'blue').
- s: Scatter

Practical Example:
---
Ek scatter plot banate hain jisme colors aur sizes add karenge:


```python

# Data
hours = [2, 4, 6, 8]
marks = [60, 75, 85, 90]

# Scatter plot with color and size
plt.scatter(hours, marks, color='green', s=100)

# Labels aur title
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Marks vs Study Hours')
plt.show()
```

Thodi Twist: Har point ka alag size aur color dena
--

```python
sizes = [50, 100, 150, 200]  # Har dot ka size alag
colors = ['red', 'blue', 'green', 'purple']  # Har dot ka color alag

plt.scatter(hours, marks, c=colors, s=sizes)
plt.xlabel('Study Hours')
plt.ylabel('Marks')
plt.title('Marks with Custom Colors and Sizes')
plt.show()
```

Example :
---

```python
days = [1, 2, 3, 4]
sales = [20, 35, 50, 45]


sizes = [30, 600, 90, 120]
colors = ['blue', 'yellow', 'red', 'green']

plt.scatter(days, sales, c=colors, s=sizes)
plt.xlabel('Days')
plt.ylabel('Sales')
plt.title('Daily Sales with Varied Styles')
plt.show()

```

Grid Aur Legend Add Karna
--
Ab hum seekhenge ki graph mein grid aur legend kaise add karte hain. Yeh graph ko aur readable aur professional banata hai.

Kyun Zaroori Hai?
--

- Grid: Graph pe lines add karta hai jo values ko samajhne mein madad karti hain—jaise scale ya ruler ki tarah.
- Legend: Jab ek graph mein multiple lines ya points hote hain, toh legend batata hai ki kaunsa color ya style kya represent karta hai. Machine learning mein comparisons ke liye useful hai.
  
Kaise Karte Hain?
----

- plt.grid(): Grid lines add karta hai.
- plt.legend(): Legend box add karta hai (pehle plot mein label dena zaroori hai).
  

Practical Example:
--
Data Kya Hai?
---
Maan lo ek chhoti dukaan hai jo Apple aur Banana bechti hai. Hum 4 din ke sales ka data track kar rahe hain:


- days: Yeh 4 din hain—Day 1, Day 2, Day 3, aur Day 4. Yeh X-axis pe jayega kyunki time ke saath sales dekhna hai.
- apple_sales: Yeh Apple ki roz ki bechi gayi quantity hai—Day 1 pe 20, Day 2 pe 25, Day 3 pe 30, Day 4 pe 35. Matlab Apple ki sale har din badh rahi hai.
- banana_sales: Yeh Banana ki roz ki bechi gayi quantity hai—Day 1 pe 15, Day 2 pe 20, Day 3 pe 25, Day 4 pe 30. Banana ki sale bhi badh rahi hai, lekin Apple se thodi kam.

Kyun Yeh Data?
  ---

Yeh data isliye liya taaki hum Apple aur Banana ke sales ko compare kar sakein aur dekhein kaunsa fruit zyada popular hai. Graph se yeh visually samajh aayega.

```python
import matplotlib.pyplot as plt

# Data
days = [1, 2, 3, 4]  # 4 din
apple_sales = [20, 25, 30, 35]  # Apple ki sales har din
banana_sales = [15, 20, 25, 30]  # Banana ki sales har din

# Plots banayein
plt.plot(days, apple_sales, label='Apple Sales', c='black')  # Apple ke liye ek line
plt.plot(days, banana_sales, label='Banana Sales' , c='red')  # Banana ke liye doosri line

# Grid aur legend add karo
plt.grid(True)  # Grid lines X aur Y pe dikhegi
plt.legend()  # Legend box jo batayega Apple aur Banana ka color

# Labels aur title
plt.xlabel('Days')  # X-axis pe din
plt.ylabel('Sales')  # Y-axis pe bechi gayi quantity
plt.title('Apple vs Banana Sales')  # Graph ka naam

# Graph dikhayein
plt.show()
```

Connect Kaise Karoge?
--

- Dekho: Apple ki line har din Banana se upar hai—matlab Apple zyada bech raha hai.
- Grid se tum exact values (jaise Day 3 pe Apple 30, Banana 25) easily dekh sakte ho.
- Legend se confusion nahi hoga ki kaunsi line Apple ki hai aur kaunsi Banana ki.

Subplots (Multiple Graphs Ek Saath):
----


Ab hum seekhenge ki ek figure mein multiple graphs (subplots) kaise banate hain. Yeh real world mein kaafi use hota hai jab tumhe ek saath alag-alag cheezein compare karni hon.

Data Kya Hai?
-----------
Maan lo tum ek student ke performance ko track kar rahe ho:

- days: 4 din (Day 1 se Day 4).
- scores: Uske test scores—[85, 87, 90, 92].
- hours: Usne kitne hours padha—[2, 3, 4, 5].
- Yeh data isliye liya taaki hum ek graph mein scores aur ek mein study hours dekhein, aur dono ka relation samajh sakein.

```python

# Data
days = [1, 2, 3, 4,5,6,7]
scores = [85, 87, 90, 92, 30, 25, 10]  # Student ke scores
hours = [2, 3, 4, 5, 1, 0.5, 0]      # Study hours

# Subplots banayein (1 row, 2 columns)

#plt.subplot(1, 2, 2) ka matlab hai ki aap ek grid bana rahe ho 1 row aur 2 columns ka, aur ye second plot create kar raha hai.


plt.subplot(1, 2, 1)  # Pehla plot
plt.plot(days, scores , linestyle='-')
plt.xlabel('Days')
plt.ylabel('Scores')
plt.title('Scores Over Time')

plt.subplot(1, 2, 2)  # Doosra plot
plt.plot(days, hours)
plt.xlabel('Days')
plt.ylabel('Hours')
plt.title('Study Hours Over Time')

# Layout adjust karo aur show
plt.tight_layout() #tight_layout() se graphs ke beech space thik rehta hai.
plt.show()
```

Data: Ek shop ke 4 din ke sales aur customers:

```python
days = [1, 2, 3, 4]
sales = [20, 25, 30, 35]    # Roz ki sales
customers = [5, 7, 10, 8]   # Roz ke customers


plt.subplot(1,2,1)
plt.plot(days,sales, linestyle='--')
plt.xlabel('Days')
plt.ylabel('Sales')

plt.title("Sales over Days")

plt.subplot(1,2,2)
plt.plot(days, customers, linestyle='--')
plt.xlabel("Days")
plt.ylabel("Customers")

plt.title("Customers Over Days")


plt.tight_layout()
plt.show()
```

chalo abb iss data ko analyse karte hain :
---

1.Sales aur Customers ka Trend Samajhna
--------
    Agar aap graphs ko properly analyze karein, to aap 3 cheezein samajh sakte hain:


1️⃣ Kya Sales Badh Rahi Hai ya Gir Rahi Hai?
----
- ✅ Agar graph upar ja raha hai, to iska matlab sales badh rahi hai.
- ✅ Agar graph neeche aa raha hai, to iska matlab sales gir rahi hai.
- ✅ Agar line zig-zag hai, to sales irregular hai, kabhi badh rahi hai, kabhi gir rahi hai.

👀 Example:
---
Agar aapke graph me days badhne ke saath sales bhi badh rahi hai, to iska matlab business grow ho raha hai.

Lekin agar end me sales neeche aa rahi hai, to iska matlab kuch issue ho sakta hai—shayad customers kam ho rahe hain ya demand gir rahi hai.



2️⃣ Customers ka Pattern Kya Hai?
---
- ✅ Kya customers har din badh rahe hain ya fluctuating hai?
- ✅ Sales aur customers ka relation hai ya nahi?

👀 Example:
--
Agar customers badh rahe hain, par sales nahi badh rahi, to iska matlab:

- Customers aa rahe hain, par shayad zyada kharidari nahi kar rahe.
- Ya phir chhoti chhoti purchases ho rahi hain.
- Agar customers kam ho rahe hain, aur sales bhi gir rahi hai, to iska matlab:

Aapke business ka footfall bhi kam ho raha hai.





3️⃣ Koi Spike ya Dip Hai Jo Important Ho Sakta Hai?
--
- ✅ Agar kisi din ekdum zyada sales ho jaye (spike), to aapko ye samajhna hoga ki aisa kyu hua?
- ✅ Agar kisi din ekdum kam sales ho jaye (dip), to aapko ye analyze karna padega ki problem kya thi?

👀 Example:
--

- Agar ek particular din customers aur sales dono spike karte hain, to ho sakta hai koi special discount offer ya festival sale ho.
- Agar ek din sales gir jaye, to ho sakta hai market me competition badh gaya ho ya customers ka interest kam ho gaya ho.


Conclusion:
----
- Dono graphs ko saath dekhne se pata chalta hai ki customers aur sales ka direct relation hai ya nahi.
- Spike aur dip analyze karna important hai, taaki future me better planning ho sake.
- Agar sales kam ho rahi hai, par customers badh rahe hain, to shayad pricing ya product strategy ko change karna pade.
